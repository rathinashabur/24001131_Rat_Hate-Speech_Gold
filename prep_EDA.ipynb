```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import re

# Memuat dataset

def read_csv(file_path, encodings_to_try):
    for encoding in encodings_to_try:
        try:
            df = pd.read_csv(file_path, encoding=encoding)
            return df
        except UnicodeDecodeError:
            pass
    raise ValueError("Unable to read the CSV file with any encoding.")

encodings_to_try = ['utf-8', 'latin1', 'cp1252', 'iso-8859-1']

try:
    df_kamusalay = read_csv('new_kamusalay.csv', encodings_to_try)
except ValueError as e:
    print(e)

try:
    df_abusive = read_csv('abusive.csv', encodings_to_try)
except ValueError as e:
    print(e)

try:
    df_data = read_csv('data.csv', encodings_to_try)
except ValueError as e:
    print(e)
```

```python
# Dictionary kamusalay
dict_kamusalay = pd.Series(df_kamusalay.iloc[:, 0].values, index=df_kamusalay.index).to_dict()

# Fungsi Text Cleaning
def text_processing(text, dict_kamusalay):
    # Memecah teks menjadi kata-kata
    words = text.split()

    # Melakukan penggantian kata-kata sesuai dengan dict_kamusalay
    for i in range(len(words)):
        word = words[i]
        if word in dict_kamusalay:
            words[i] = dict_kamusalay[word]

    # Menggabungkan kembali kata-kata menjadi teks
    text = ' '.join(words)

    # Menjadi lowercase
    text = text.lower()

    # Menghilangkan punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Menghilangkan angka
    text = re.sub(r'\d+', '', text)
    
    # Menghilangkan spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()

    # Menghilangkan kata yang hanya satu huruf
    text = re.sub(r'\b[a-zA-Z]\b', ' ', text)
    return text

# Membersihkan teks tweet dan mengganti kata sesuai kamusalay
df_data_copy['Tweet'] = df_data_copy['Tweet'].apply(lambda x: text_processing(x, dict_kamusalay))

df_data_copy['Tweet']
```

```python
# _____ EDA & VISUALISASI __________

# Melihat Panjang Tweet
df_data_copy['Tweet_Length'] = df_data_copy['Tweet'].apply(len)

# Histogram distribusi panjang tweet
plt.figure(figsize=(10, 6))
sns.histplot(data=df_data_copy, x='Tweet_Length', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Tweet Length')
plt.xlabel('Tweet Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# Menghitung total persentase HS dan Abusive dan Netral
total_hs_percentage = df_data_copy['HS'].sum() / len(df_data_copy) * 100
total_abusive_percentage = df_data_copy['Abusive'].sum() / len(df_data_copy) * 100
total_netral_percentage = 100 - total_hs_percentage - total_abusive_percentage

labels = ['Hate Speech', 'Abusive', 'Neutral']
sizes = [total_hs_percentage, total_abusive_percentage, total_netral_percentage]
colors = ['#ff9999','#66b3ff','#99ff99']
explode = (0.1, 0.1, 0)  # explode the 1st and 2nd slices

plt.figure(figsize=(6, 4))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Percentage of Tweets in Different Categories')
plt.axis('equal')
plt.show()
```

```python
# Menghitung jumlah entri untuk setiap kategori
total_entries = len(df_data_copy)
hs_individual_count = df_data_copy['HS_Individual'].sum()
hs_group_count = df_data_copy['HS_Group'].sum()
hs_religion_count = df_data_copy['HS_Religion'].sum()
hs_race_count = df_data_copy['HS_Race'].sum()
hs_physical_count = df_data_copy['HS_Physical'].sum()
hs_gender_count = df_data_copy['HS_Gender'].sum()
hs_other_count = df_data_copy['HS_Other'].sum()

# Menghitung persentase untuk setiap kategori
hs_individual_percentage = hs_individual_count / total_entries * 100
hs_group_percentage = hs_group_count / total_entries * 100
hs_religion_percentage = hs_religion_count / total_entries * 100
hs_race_percentage = hs_race_count / total_entries * 100
hs_physical_percentage = hs_physical_count / total_entries * 100
hs_gender_percentage = hs_gender_count / total_entries * 100
hs_other_percentage = hs_other_count / total_entries * 100

# Menyusun data untuk pie chart
labels = ['HS_Individual', 'HS_Group', 'HS_Religion', 'HS_Race', 'HS_Physical', 'HS_Gender', 'HS_Other']
sizes = [hs_individual_percentage, hs_group_percentage, hs_religion_percentage, hs_race_percentage, hs_physical_percentage, hs_gender_percentage, hs_other_percentage]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6','#c4e17f']
explode = (0.1, 0, 0, 0, 0, 0, 0)

# Membuat pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Percentage of Hate Speech Categories')
plt.axis('equal')
plt.show()
```
